```python
# 코드 9-28 라이브러리 호출
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.python.data import Dataset
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
# 코드 9-29 데이터셋 로딩 및 모델 훈련
df = pd.read_csv("data/covtype.csv")
x = df[df.columns[:54]]
y = df.Cover_Type  # 정답(레이블)을 Cover_Type 칼럼으로 지정

# 훈련과 검증 데이터셋으로 분리하며, 전체 데이터셋 중 70%를 훈련용으로 사용
x_train, x_test, y_train, y_test = train_test_split(
    x, y, train_size=0.7, random_state=90
)

model = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(64, activation="relu"),
        # 출력층은 소프트맥스 활성 함수 사용
        tf.keras.layers.Dense(8, activation="softmax"),
    ]
)

# y가 다중 분류가 가능한 값을 갖기 때문에 sparse-categorical crossentropy 손실 함수 사용
model.compile(
    optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"]
)

# 모델 훈련
history1 = model.fit(
    x_train, y_train, epochs=26, batch_size=60, validation_data=(x_test, y_test)
)
```

```python
# 코드 9-30 데이터 정규화
from sklearn import preprocessing

df = pd.read_csv("data/covtype.csv")
x = df[df.columns[:55]]
y = df.Cover_Type
x_train, x_test, y_train, y_test = train_test_split(
    x, y, train_size=0.7, random_state=90
)

# 훈련 데이터셋에서 정규화가 필요한 칼럼 선택
train_norm = x_train[x_train.columns[:10]]
# 검증 데이터셋에서 정규화가 필요한 칼럼 선택
test_norm = x_test[x_test.columns[:10]]

# 훈련 데이터셋 정규화
std_scale = preprocessing.StandardScaler().fit(train_norm)


x_train_norm = std_scale.transform(train_norm)
# 넘파이(numpy) 배열을 데이터프레임(DataFrame)으로 변환
training_norm_col = pd.DataFrame(
    x_train_norm, index=train_norm.index, columns=train_norm.columns
)
x_train.update(training_norm_col)
print(x_train.head())


x_test_norm = std_scale.transform(test_norm)  # 검증 데이터셋 정규화
testing_norm_col = pd.DataFrame(
    x_test_norm, index=test_norm.index, columns=test_norm.columns
)
x_test.update(testing_norm_col)
print(x_test.head())
```

```python
# 코드 9-31 데이터셋 로딩 및 모델 훈련
model = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(8, activation="softmax"),
    ]
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

history2 = model.fit(
    x_train, y_train, epochs=26, batch_size=60, validation_data=(x_test, y_test)
)
```
